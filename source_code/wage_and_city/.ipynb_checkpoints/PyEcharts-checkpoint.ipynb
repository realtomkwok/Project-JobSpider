```python
from pyecharts.charts import Bar
import pandas as pd

df = pd.read_excel('C#.xlsx')
df.experience.value_counts()
data = df.groupby(['city','experience']).avgWage.mean().round(2).reset_index(name='salary')
data
```

```python
data.iloc[12:16]
```

data.iloc

```python
shanghai = list(data.iloc[0:4].salary)
beijing = list(data.iloc[4:8].salary)
guangzhou = list(data.iloc[8:12].salary)
shenzhen = list(data.iloc[12:16].salary)
```

```python
bar = Bar()
bar.add_xaxis(['1-5年', '10年以上', '5-10年', '不限'])
bar.add_yaxis('上海', shanghai)
bar.add_yaxis('北京', beijing)
bar.add_yaxis('广州', guangzhou)
bar.add_yaxis('深圳', shenzhen)
bar.render_notebook()
```
